```python
'''
Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.

Ans-The find() method is used to query the MongoDB database and retrieve documents from a specific collection that match a specified set of criteria. This method can be used with various query options to filter, sort, and limit the results.

Here's an example of how you can use the find() method in Python to query a MongoDB database:

'''

import pymongo

client=pymongo.MongoClient('mongodb+srv://atul:atul@cluster0.vxsgpie.mongodb.net/?retryWrites=true&w=majority')
db=client["atul"]

data=[
        {  "_id": 1, "name": "John Doe","age": 25,"gpa": 8.5 },

        {  "_id": 2,"name": "smith","age": 22,"gpa": 3.5 },

        {  "_id": 3,"name": "jonson","age": 28,"gpa": 7.5 }
 ]
coll_result=db['result']
coll_result.insert_many(data)
for i in coll_result.find():
    print(i)

    
for i in coll_result.find({'_id':{'$gt':2}}):
    print(i)

```

```python

'''Q7. Explain why delete_one(), delete_many(), and drop() is used.

Ans-In MongoDB, delete operations are used to remove documents from a collection. The three main methods for deleting documents are delete_one(), delete_many(), and drop(). Here's a brief explanation of when and why each method is used:

delete_one(): This method is used to delete a single document from a collection that matches a specified filter. If multiple documents match the filter, only the first one found will be deleted. This method is useful when you want to delete a specific document or a small number of documents from a collection.

delete_many(): This method is used to delete all documents from a collection that match a specified filter. This method is useful when you want to delete multiple documents from a collection at once. It can also be used to delete all documents in a collection by passing an empty filter object.

drop(): This method is used to completely remove a collection from a database. This method is useful when you want to delete an entire collection and all its documents. It is a more drastic method of deleting data than delete_one() or delete_many(), and should be used with caution.

In summary, delete_one() is used to remove a single document that matches a filter, delete_many() is used to remove multiple documents that match a filter, and drop() is used to delete an entire collection.
'''


import pymongo
client=pymongo.MongoClient('mongodb+srv://atul:atul@cluster0.vxsgpie.mongodb.net/?retryWrites=true&w=majority')
db=client['atul']

collection=db['my_record']

query = { "name": "atul" }

collection.delete_one(query)

query = { "name": "Ankit" }
collection.delete_many(query)

collection.drop()
```
